# 无图例的散点图

```python
def scatter_plot1(datingData,datingLabels):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.scatter(datingData[:,1],datingData[:,2],15.0*np.array(datingLabels),15.0*np.array(datingLabels))
    #第一个15.0*np.array(datingLabels)表示点的大小不一样，第二个15.0*np.array(datingLabels)表示颜色不同
    ax.axis([-2,25,-0.2,2.0])
    plt.show()
#应该要调用函数才能显示
```

```python
scatter_plot1(datingData,datingLabels)
```

# 有图例的散点图

```python
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"c:\windows\fonts\SimSun.ttc", size=14)
def scatter_plot2(datingData,datingLabels):
    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)
    datingLabels = np.array(datingLabels)
    idx_1 = np.where(datingLabels==1)
    p1 = ax.scatter(datingData[idx_1,0],datingData[idx_1,1],marker = '*',color = 'r',label='1',s=10)
    idx_2 = np.where(datingLabels==2)
    p2 = ax.scatter(datingData[idx_2,0],datingData[idx_2,1],marker = 'o',color ='g',label='2',s=20)
    idx_3 = np.where(datingLabels==3)
    p3 = ax.scatter(datingData[idx_3,0],datingData[idx_3,1],marker = '+',color ='b',label='3',s=30)
    plt.xlabel("每年获取的飞行里程数",fontproperties=font)
    plt.ylabel("玩视频游戏所消耗的事件百分比",fontproperties=font)
    ax.legend((p1,p2,p3),("不喜欢","魅力一般","极具魅力"),loc=2,prop=font)
    plt.show()
```

```python
scatter_plot2(datingData,datingLabels)
```
