```python
def datingClassTest():
    hoRatio = 0.1  #用整个数据的百分之多少来进行测试
    datingDataMat,datingLabels = filematrix('datingTestSet2.txt')#读取数据
    normMat,ranges,minVals = autoNorm(datingDataMat)#读取处理好的特征矩阵
    m = normMat.shape[0]#得到特征矩阵的行数
    print(m)
    numTestVecs = int(m*hoRatio)#用来测试的特征矩阵的行数
    print(normMat)
    print(np.shape(normMat))
    errorCount = 0.0
    for i in range(numTestVecs):
        classifierResult = classify(normMat[i, :], normMat[numTestVecs:m, :], datingLabels[numTestVecs:m], 3)
        #利用前numTestVecs行数据进行测试，利用从numTestVecs行数据进行测试到m行数据进行测试
        #classify(inX,dataSet,labels,k)
        print("the classifier came back with: %d, the real answer is: %d" % (classifierResult, datingLabels[i]))
        if (classifierResult != datingLabels[i]): errorCount += 1.0 
            #如果预测结果和真实结果不一样，就在错误个数上加1
    print("the total error rate is: %f" % (errorCount / float(numTestVecs)))
    print(errorCount)
```

```python
datingClassTest()
```
